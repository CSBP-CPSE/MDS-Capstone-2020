```python
import pandas as pd
```

```python
# read in odhf file
df = pd.read_csv('odhf_v1.csv')
```

```python
# we see there are no missing values for province
df.info()
df.head()
```

```python
# list and count number of unique provinces, OK
df['province'].value_counts()
```

```python
# filter df for facilities in ON
df_on = df[df['province'].isin(['on'])]
df_on.info()
df_on.head()

# there are 3 types of ODHF facilities Nursing and residential care facilities, Ambulatory health care services and Hospitals                                   
df_on['odhf_facility_type'].value_counts() 

# there is one duplicate facility name
df_on['facility_name'].nunique() 

```

```python
df_on.to_csv(r'ohdf_v1_ontario.csv', index = False)
```
