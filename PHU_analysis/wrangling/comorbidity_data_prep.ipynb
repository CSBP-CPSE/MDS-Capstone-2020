```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
# Import the comorbidity data
df = pd.read_csv('13100113-eng/13100113.csv')

df.info()
df.head()
```

```python
# Filter for Ontario data
df_on = df[df['GEO'].str.contains('Ontario', regex=False, na=False)]

# Remove data for the entire province of Ontario
df_on.drop(df.index[df['GEO'] == 'Ontario'], inplace = True)

# Create a new column with just the PHU name
df_on['phu_name'] = df['GEO'].str.split(',').str[0]

# Create a new column with the HR_UID
df_on['HR_UID'] = df.DGUID.str[9:]

df_on.info()
df_on.head()
```

```python
# Subset the data
subset = df_on.loc[df_on['Age group'] == 'Total, 12 years and over']
subset = subset[subset.REF_DATE == '2017/2018']
subset = subset[subset['Sex'] == 'Both sexes' ]

listofchar = ['Low 95% confidence interval, number of persons', 
              'High 95% confidence interval, number of persons', 
              'Low 95% confidence interval, percent',
              'High 95% confidence interval, percent', 
              'Statistically different from the Canada rate', 
              'Statistically different from the provincial rate', 
              'Statistically different from previous reference period']

for each in listofchar:
    subset = subset[subset.Characteristics != each]

subset.head()
```

```python
# Drop columns that are not needed
listofcols = ['REF_DATE', 'GEO', 'DGUID', 'Age group', 'Characteristics', 'Sex', 'UOM_ID',
              'SCALAR_FACTOR', 'SCALAR_ID', 'VECTOR', 'COORDINATE', 'STATUS', 'SYMBOL', 
              'TERMINATED', 'DECIMALS']

subset.drop(listofcols, axis = 1, inplace = True)
subset.info()
subset.head()
```

```python
# Subset data, just numbers, and pivot
df_num = subset[subset.UOM == 'Number']
df_num = df_num.pivot_table('VALUE', ['phu_name', 'HR_UID'], 'Indicators')
df_num.info()
df_num.head(5)
```

```python
# Subset data, just percents, and pivot
df_percent = subset[subset.UOM == 'Percent']
df_percent = df_percent.pivot_table('VALUE', ['phu_name', 'HR_UID'], 'Indicators')
df_percent.info()
df_percent.head(5)
```

```python
df_percent = df_percent.reset_index()
df_num = df_num.reset_index()

# Export the comorbidities data
df_percent.to_csv(r'comorbidity_data_percent.csv', index = False)
df_num.to_csv(r'comorbiditiy_data_num.csv', index = False)
```
