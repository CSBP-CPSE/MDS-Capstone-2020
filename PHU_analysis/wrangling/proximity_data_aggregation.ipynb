```python
# Import PHU assignments for each db using only the necessary columns
db_phu = pd.read_csv('Joined_DB_to_PHU.csv', usecols=([1, 28, 29, 30]))
len(set(db_phu['DBUID'])) # There are 133,214 unique DBs 
len(set(db_phu['HR_UID'])) # There are 36 unique HRs 
db_phu.info()
db_phu.head()

```

```python
# Import the proximity data for aggregation to the PHU level
prox = pd.read_csv('proximity_data_prep_forQGIS.csv')
prox.info()
len(set(prox['DBUID'])) # There are 132,711 DBs with proximity data
prox.head()
```

```python
# Check that all DBs in proximity data are present in the DB to PHU data
prox['DBUID'].isin(db_phu['DBUID']).value_counts() # all DBs present in the proximity data are represented
```

## PHU assignments

```python
# Join DB to PHU assignments to proximity data
merged = pd.merge(left=prox, right=db_phu, how='left', left_on='DBUID', right_on='DBUID')

# Convert population columns to numeric types
merged[['DBPOP']] = merged[['DBPOP']].apply(pd.to_numeric) 

merged.info()
merged.head(5)
```

## Aggregate proximity measures from DB to PHU level

```python
# Create a new column which contains the PHU population number for that db
merged['HRPOP'] = merged.groupby('HR_UID')['DBPOP'].transform('sum')

# Create a new column which is the ratio of the DBPOP/HRPOP
merged['DBHRPOP'] = merged['DBPOP']/merged['HRPOP']

# Inspect the results
merged.info()
# merged['HRPOP'].value_counts()
# merged['HRPOP'].describe()
# # Merged['DBPOP'].value_counts() # 19599 zeros in DBPOP
# merged['DBPOP'].describe()
# # Merged['DBHRPOP'].value_counts() # # 19599 zeros in DBHRPOP
# merged['DBHRPOP'].describe()
```

### 1. Aggregate the proximity measures by taking the sum of population weighted measures for each HR

```python
# Create a list containing prox measures column names
list_prox = ['prox_idx_emp', 'prox_idx_pharma', 'prox_idx_childcare',
             'prox_idx_health', 'prox_idx_grocery', 'prox_idx_educpri',
             'prox_idx_educsec', 'prox_idx_lib', 'prox_idx_parks', 
             'prox_idx_transit']

# Create new weighted proximity columns for each row (ie. DBHRPOP * prox_)
for each in list_prox:
    merged[each + '_wt'] = merged[each] * merged['DBHRPOP']
    
# Create a list with weighted prox measures column names
list_prox_wt = [item + '_wt' for item in list_prox]
print(list_prox_wt)
    
# list_prox_wt

# Create a new PHU level df where weighted proximity values are summed
phu_prox_wt = merged.groupby(['HR_UID', 'ENG_LABEL', 'FRE_LABEL'])[list_prox_wt].sum()
phu_prox_wt.info()
phu_prox_wt.head()
```

### 2. Aggregate the proximity measures by taking the median of the measures for each HR

```python
# Create list of column nammes for the median proximity measures
list_prox_med = []
for each in list_prox:
    list_prox_med.append(each + '_med')
# list_prox_med
```

```python
list_prox2 = ['prox_idx_emp', 'prox_idx_pharma', 'prox_idx_childcare',
             'prox_idx_health', 'prox_idx_grocery', 'prox_idx_educpri',
             'prox_idx_educsec', 'prox_idx_lib', 'prox_idx_parks', 
             'prox_idx_transit', 'HRPOP']


phu_prox_med = merged.groupby(['HR_UID', 'ENG_LABEL', 'FRE_LABEL'])[list_prox2].median()

# Rename columns
dictionary = dict(zip(list_prox, list_prox_med))
# dictionary
phu_prox_med.rename(columns = dictionary, inplace=True)
phu_prox_med.info()
phu_prox_med.head()
```

### 3. Aggregate the amenity dense measure by taking a mean

```python
# Create a df with average amenity dense measure by PHU
phu_amenity = merged.groupby(['HR_UID', 'ENG_LABEL', 'FRE_LABEL'])['amenity_dense'].mean()
phu_amenity = phu_amenity.to_frame()
phu_amenity.info()
phu_amenity.head()
```

## Merge and export the aggregated proximity data

```python
# Merge aggregated proximity data (weighted proximity, median proximity and amenity)
phu_prox_full = phu_amenity.merge(phu_prox_wt, how='outer', 
                                  left_on=['HR_UID', 'ENG_LABEL', 'FRE_LABEL'], 
                                  right_on=['HR_UID', 'ENG_LABEL', 'FRE_LABEL']).merge(phu_prox_med, 
                                                                          how='outer', 
                                                                          left_on=['HR_UID', 'ENG_LABEL', 'FRE_LABEL'], 
                                                                          right_on=['HR_UID', 'ENG_LABEL', 'FRE_LABEL'])
phu_prox_full = phu_prox_full.reset_index()
phu_prox_full.info()
phu_prox_full.head()
```

```python
# Export the PHU level proximity data
phu_prox_full.to_csv(r'proximity_data_aggregated.csv', index = False)
```
