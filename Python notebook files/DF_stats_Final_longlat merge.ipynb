# Merging

```python
new_df = pd.merge(final_data, latlong, how = 'left', on = 'address')
```

```python
new_df.head(10)
```

```python
# Checking the empty columns entires
new_df[new_df['latitude_y'].notnull()]
```

```python
import numpy as np

new_df['lati'] = np.nan
new_df['long'] = np.nan

new_df['lati'] = np.where((new_df.latitude_y.isnull()), new_df['latitude_x'], new_df['latitude_y'])
new_df['long'] = np.where((new_df.longitude_y.isnull()), new_df['longitude_x'], new_df['longitude_y'])
```

```python
new_df.drop(['latitude_x', 'latitude_y', 'longitude_x','longitude_y'], axis=1,inplace=True)
```
