```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
# Import the comorbidity data
df = pd.read_csv('comorbidity_data.csv')

# Subset the DGUID so that it matches
df['ts_guid'] = df.DGUID.str[5:]
df['HRUID'] = df.DGUID.str[9:]

df.info()
df.head()
```

```python
# Obtain DGUIDs for each health region in Ontario
df2 = pd.read_csv('hbp-percent.csv')
df2.info()
df2 = df2.loc[df2['Province/Territory'] == 'Ontario']
df2.drop(['Province/Territory', 'hbp-percent', 'Data Comment'], axis = 1, inplace = True)

# Subset the DGUID so that it matches
df2['ts_guid'] = df2.DGUID.str[5:]
df2
```

```python
df3 = pd.merge(left = df2, right = df, how = 'left', left_on = 'ts_guid', right_on = 'ts_guid')
df3.info()
df3.head()
```

```python
# Subset the data
df3 = df3.loc[df3['Age group'] == 'Total, 12 years and over']
df3 = df3[df3.REF_DATE == '2017/2018']
df3 = df3[df3['Sex'] == 'Both sexes' ]

listofchar = ['Low 95% confidence interval, number of persons', 
              'High 95% confidence interval, number of persons', 
              'Low 95% confidence interval, percent',
              'High 95% confidence interval, percent', 
              'Statistically different from the Canada rate', 
              'Statistically different from the provincial rate', 
              'Statistically different from previous reference period']

for each in listofchar:
    df3 = df3[df3.Characteristics != each]

df3.head()
```

```python
# Drop columns that are not needed
listofcols = ['DGUID_x', 'GEO', 'Age group', 'Sex', 
              'SCALAR_FACTOR', 'SCALAR_ID', 'STATUS', 'SYMBOL', 
              'TERMINATED', 'DECIMALS', 'VECTOR', 'COORDINATE', 'UOM_ID']

df3.drop(listofcols, axis = 1, inplace = True)
```

```python
df3.head()
```

```python
# Subset data, just numbers, and pivot
df_num = df3[df3.Characteristics == 'Number of persons']
df_num = df_num.pivot_table('VALUE', ['Location', 'ts_guid', 'REF_DATE', 'DGUID_y', 'HRUID'], 'Indicators')
df_num.head(10)
# df_num.info()
```

```python
# Subset data, just percents, and pivot
df_percent = df3[df3.Characteristics == 'Percent']
df_percent.head()
df_percent = df_percent.pivot_table('VALUE', ['Location', 'ts_guid', 'REF_DATE', 'DGUID_y', 'HRUID'], 'Indicators')
df_percent.head(10)
# df_percent.info()
```

```python
df_percent = df_percent.reset_index()
df_num = df_num.reset_index()
df_percent.head()
df_percent.info()

# Export the comorbidities data
df_percent.to_csv(r'comorbidities_percent.csv', index = False)
df_num.to_csv(r'comorbidities_num.csv', index = False)
```
